## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
# Research Question

# Which content pages should be prioritized for human review based on observable search, content, and engagement signals?

# Unit of Analysis

# The unit of analysis is a content page. Each page will be evaluated using available signals such as impressions, clicks, CTR, average position, content age, freshness, and engagement.

# Decision

# The system should help decide which pages should be reviewed first when the content team has limited time and cannot review every page.

# Output

# The expected output is a ranked list of pages, where higher-ranked pages represent stronger candidates for review. The recommendations should also include reason codes explaining why a page was prioritized.

# Action

# Based on the recommendation and human review, a content team could decide to refresh, expand, protect, monitor, or potentially prune a page.

# Cost of a Wrong Recommendation

# A false positive could waste time and resources by prioritizing a page that did not need attention. A false negative could be more costly because the system might miss a page with an important opportunity or genuine decline.

# Why Data or ML Can Help

# Reviewing a large number of pages manually is difficult. Data and machine learning can combine multiple observable signals to help prioritize pages more efficiently. The starter results also suggest that a learned ranking may outperform a simple fixed rule. However, the model should support human decisions rather than automatically deciding that a page must be changed.
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

# Load the starter dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total pages in dataset:", len(df))

# Pages showing a declining trend
declining_pages = (df["trend_direction"] == "down").sum()
declining_pct = (declining_pages / len(df)) * 100

print("Pages with declining trend:", declining_pages)
print(f"Percentage of declining pages: {declining_pct:.2f}%")

# Declining pages that still have meaningful search demand
declining_with_demand = (
    (df["trend_direction"] == "down") &
    (df["impressions_90d"] >= 100)
).sum()

print(
    "Declining pages with at least 100 impressions:",
    declining_with_demand
)
```
